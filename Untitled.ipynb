```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("uber.csv")
```

```python
data["Date/Time"] = data["Date/Time"].map(pd.to_datetime)
```

```python
data.head()
```

```python
data["Day"] = data["Date/Time"].apply(lambda x: x.day)
data["Weekday"] = data["Date/Time"].apply(lambda x: x.weekday())
data["Hour"] = data["Date/Time"].apply(lambda x: x.hour)
```

```python
print(data.head())
```

```python
sns.set(rc={'figure.figsize': (12,10)})
sns.distplot(data["Day"]).set(title='Uber Rides Day-to-Day Density')
```

```python
sns.distplot(data["Hour"])
```

```python
sns.distplot(data["Weekday"])
```

```python
df = data.groupby(["Weekday","Hour"]).apply(lambda x: len(x)) # Grouping Weekdays and Hours and getting the length of elements of uber rides
df = df.unstack()
sns.heatmap(df, annot=False)
```

```python
data.plot(kind='scatter', x='Lon',y='Lat', alpha=0.4, s=data['Day'],label='Uber Trips',figsize=(12,8),cmap=plt.get_cmap('jet'))
plt.title("Uber Trip Analysis in New York")
plt.legend()
plt.show()
```
